```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
k = int(input("Enter the value of the strike price: "))#strike price
p = int(input("Enter the value of the premium : "))#premium
s = np.linspace(k-50,k+50,100)
```

```python
#buy call
def buy_call(s,k,p):
    return np.maximum(0,s-k)-p
#sell call
def sell_call(s,k,p):
    return p-np.maximum(0,s-k)
#buy put
def buy_put(s,k,p):
    return np.maximum(0,k-s)-p
#sell put
def sell_put(s,k,p):
    return p-np.maximum(0,k-s)
```

```python
buy_call = buy_call(s,k,p)
sell_call = sell_call(s,k,p)
buy_put = buy_put(s,k,p)
sell_put = sell_put(s,k,p)
```

```python
plt.plot(s,buy_call,label='buy call')
plt.plot(s,sell_call,label='sell call')
plt.plot(s,buy_put,label='buy put')
plt.plot(s,sell_put,label='sell put')
plt.legend()
plt.xlabel('Stock Price')
plt.ylabel('Profit/Loss')
plt.show()
print(f"The buy call constal loss of {p} before strike price ={k} then atmost profit is {np.max(buy_call)}")
print(f"The sell call constal profit of {p} before strike price ={k} then it atmost loss is {np.min(sell_call)}")
print(f"The buy put constal loss  of {p} after strike price ={k} before atmost profit is {np.max(buy_put)}")
print(f"The sell put constal of {p} after strike price ={k} then atmost loss is {np.min(sell_put)}")
print(f"The price  profit-loss = 0 is {k+p} and {k-p}")
print(f"The price when the profit or loss change after {k}")
```
